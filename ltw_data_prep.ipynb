Add year column and state number to dataframes

```python
dfs = [by08, by13, by18, he09, he13, he18, nrw10, nrw12, nrw17, ns08, ns13, ns17, th09, th14, th19]
years = [2008, 2013, 2018, 2009, 2013, 2018, 2010, 2012, 2017, 2008, 2013, 2017, 2009, 2014, 2019]
states = ['BY', 'BY', 'BY', 'HE', 'HE', 'HR', 'NR', 'NW', 'NW', 'NI', 'NI', 'NI',  'TH', 'TH', 'TH']
for df, year, state in zip(dfs, years, states):
    df.insert(loc=0, column='year', value=year)
    df.insert(loc=0, column='Land', value=state)
```

Concatenate dataframes

```python
dfs = [by08, by13, by18, he09, he13, he18, nrw10, nrw12, nrw17, ns08, ns13, ns17, th09, th14, th19]
ltw = pd.concat(dfs, sort=False)
```

Check State Abbreviations

```python
ltw['Land'].unique()
```

State abbreviation is different in he18 and nrw10 --> change

```python
ltw['Land'] = np.where(ltw['Land'] == 'NR', 'NW', ltw['Land']) 
ltw['Land'] = np.where(ltw['Land'] == 'HR', 'HE', ltw['Land']) 
ltw['Land'].unique()
```

Fix the AGS

```python
ltw = ltw[ltw['AGS'] != '']
ltw['AGS'] = ltw['AGS'].astype(int)
```

```python
ltw['AGS'] = ltw['AGS'].astype(str)
ltw['AGS'] = np.where(ltw['AGS'].str.len() <= 2, ltw['AGS'] + '0'*6, ltw['AGS']) # add 6*0 after AGS for states with 1 or 2 digit AGS
ltw['AGS'] = np.where(ltw['AGS'].str.len() < 6, ltw['AGS'] + '0'*3, ltw['AGS']) # add trailing 000 for kreisfreie Städte (AGS has 4-5 digits before, 7-8 after)
ltw['AGS'] = np.where(ltw['AGS'].str.len() < 8, '0' + ltw['AGS'], ltw['AGS']) # add leading 0 for state ids < 10 (AGS has 7 digits)
ltw.shape
```

Check if AGS was properly calculated

```python
ltw_ags = ltw[ltw['AGS'].str.len() == 8]
ltw_ags.shape
# --> all AGS are 8 digits long
# Still there are some duplicates, e.g. Gießen Stadt and Landkreis --> these won't be matched when merging with treatment data
```

Add County ID

```python
ltw['Kreis'] = ltw['AGS'].str.slice(0,5)
```

Add state ID

```python
ltw['state_ID'] = ltw['AGS'].str.slice(0,2)
```

Clean data and sum votes for AfD and Andere

```python
ltw.rename(columns = {'GŸltige Stimmen':'Gültig', 'GrŸne': 'Grüne'}, inplace = True)
andere = ['AfD', 'Sonstige Parteien']
ltw['Andere'] = ltw[andere].sum(axis=1)
ltw.head()
```

Calculate vote shares in percent

```python
parties = ['Union',	'SPD', 'FDP', 'Linke', 'Grüne', 'Andere']
for party in parties:
    ltw.loc[:, party] = (ltw[party] / ltw['Gültig'])*100
ltw.head(10)
```

Add post-treatment indicator

```python
ltw['post_2014'] = np.where(ltw['year'] >= 2014, 1, 0)
ltw['post_2010'] = np.where(ltw['year'] > 2010, 1, 0)
ltw = ltw.sort_values(['AGS', 'year'])
ltw.head(10)
```

Clean dataset

```python
ltw = ltw[['AGS', 'Land', 'state_ID', 'Kreis', 'year', 'post_2010', 'post_2014', 'Union', 'SPD', 'Grüne', 'FDP', 'Linke', 'Andere']]
ltw.shape
```

Drop obs. with missing values for election results

```python
ltw = ltw.dropna(axis='rows', how='any', subset=['Union', 'SPD', 'FDP', 'Linke', 'Grüne', 'Andere'])
ltw.shape
```

Merge with treatment dataset

```python
treatment = pd.read_csv(f'{path}/data/treatment.csv', converters = {'AGS': str})
ltw_treat = pd.merge(ltw, treatment, on='AGS', how='outer', indicator=True)
ltw_treat.head(10)
```

Inspect merge results

```python
print(ltw_treat['_merge'].describe())
```

Clean and export

```python
ltw_treat = ltw_treat[ltw_treat['_merge'] == 'both']
ltw_treat = ltw_treat.drop(columns=['_merge'])
ltw_treat = ltw_treat.set_index(['AGS', 'GEN', 'year'])
ltw_treat.to_csv(f'{path}/data/ltw_treat.csv', encoding = 'utf-8-sig')
```

Merge in control variables

```python
ltw_treat = ltw_treat.reset_index()
controls = pd.read_csv(f'{path}/data/controls.csv', converters = {'AGS': str})
ltw_controls = pd.merge(ltw_treat, controls, on='AGS', how='outer', indicator=True)
ltw_controls.head(10)
```

```python
print(ltw_controls['_merge'].describe())
```

```python
ltw_controls = ltw_controls[ltw_controls['_merge'] == 'both']
ltw_controls = ltw_controls.set_index(['AGS', 'GEN', 'year'])
ltw_controls = ltw_controls.drop(columns=['_merge'])
```

Export dataset

```python
ltw_controls.to_csv(f'{path}/data/ltw_control.csv', encoding = 'utf-8-sig')
ltw_controls.head()
```
