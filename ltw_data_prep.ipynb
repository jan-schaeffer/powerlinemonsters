Fix AGS, add year column and state number to dataframes

```python
# Define function to fix the AGS
def ags_fixer(df):
    df = df[df['AGS'] != '']
    df['AGS'] = df['AGS'].astype(int)
    df['AGS'] = df['AGS'].astype(str)
    df['AGS'] = np.where(df['AGS'].str.len() <= 2, df['AGS'] + '0'*6, df['AGS']) # add 6*0 after AGS for states with 1 or 2 digit AGS
    df['AGS'] = np.where(df['AGS'].str.len() < 6, df['AGS'] + '0'*3, df['AGS']) # add trailing 000 for kreisfreie Städte (AGS has 4-5 digits before, 7-8 after)
    df['AGS'] = np.where(df['AGS'].str.len() < 8, '0' + df['AGS'], df['AGS']) # add leading 0 for state ids < 10 (AGS has 7 digits)
    return df
```

```python
dfs = [by08, by13, by18, he09, he13, he18, nrw10, nrw12, nrw17, ns08, ns13, ns17, th09, th14, th19]
years = [2008, 2013, 2018, 2009, 2013, 2018, 2010, 2012, 2017, 2008, 2013, 2017, 2009, 2014, 2019]
states = ['BY', 'BY', 'BY', 'HE', 'HE', 'HE', 'NW', 'NW', 'NW', 'NI', 'NI', 'NI',  'TH', 'TH', 'TH']
for i, (year, df, state) in enumerate(zip(years, dfs, states)):
    try:
        # insert year column and state
        dfs[i].insert(loc=0, column='year', value=year)
        dfs[i].insert(loc=0, column='Land', value=state)
        dfs[i] = dfs[i][dfs[i]['AGS'].str.len() > 3]
        # fix the AGS
        dfs[i]['AGS'] = dfs[i]['AGS'].astype(str)
        dfs[i]['AGS'] = np.where(dfs[i]['AGS'].str.len() < 6, dfs[i]['AGS'] + '0'*3, dfs[i]['AGS']) # add trailing 000 for kreisfreie Städte (AGS has 4-5 digits before, 7-8 after)
        dfs[i]['AGS'] = np.where(dfs[i]['AGS'].str.len() < 8, '0' + dfs[i]['AGS'], dfs[i]['AGS']) # add leading 0 for state ids < 10 (AGS has 7 digits)
    except Exception:
        pass
    # don't have to change the AGS for 2013
    if year == 2013:
        continue
    # read in AGS transition datasets for same year and previous year
    ags_trans_same = pd.read_csv(f'{path}/data/ags_transition.csv', usecols=[f'AGS_{year}', 'AGS_2013'], converters={f'AGS_{year}': str, 'AGS_2013': str})
    ags_trans_same.rename(columns = {f'AGS_{year}':'AGS_new'}, inplace = True)
    ags_trans_prev = pd.read_csv(f'{path}/data/ags_transition.csv', usecols=[f'AGS_{year-1}', 'AGS_2013'], converters={f'AGS_{year-1}': str, 'AGS_2013': str})
    ags_trans_prev.rename(columns = {f'AGS_{year-1}':'AGS_new'}, inplace = True)
    ags_trans = pd.concat([ags_trans_same, ags_trans_prev])
    # merge dfs[i] with ags trans
    dfs[i] = dfs[i].merge(ags_trans, left_on='AGS', right_on=f'AGS_new', how='left', indicator=True)
    print(dfs[i]._merge.value_counts())
    # if merged successfully, replace AGS
    dfs[i]['AGS'] = np.where(dfs[i]['AGS_2013'].notna(), dfs[i]['AGS_2013'], dfs[i]['AGS'])
    dfs[i] = dfs[i].drop(['AGS_2013', f'AGS_new'], axis=1)
    # move AGS to front
    AGS = dfs[i].pop('AGS')
    dfs[i].insert(1, 'AGS', AGS)
# Concatenate dfs
ltw = pd.concat(dfs, sort=False)
ltw.shape
```

```python
ltw = ltw.drop_duplicates(keep='first')
ltw.shape
```

Clean data and sum votes for AfD and Andere

```python
ltw.rename(columns = {'GŸltige Stimmen':'Gültig', 'GrŸne': 'Grüne'}, inplace = True)
andere = ['AfD', 'Sonstige Parteien']
ltw['Andere'] = ltw[andere].sum(axis=1)
ltw.head()
```

```python
parties = ['Union',	'SPD', 'FDP', 'Linke', 'Grüne', 'Andere']
for party in parties:
    ltw.loc[:, party] = (ltw[party] / ltw['Gültig'])*100
ltw.head(10)
```
