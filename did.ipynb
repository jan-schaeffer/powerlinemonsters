First, import needed modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import os
```

Initalize variables

```python
path = os.getcwd()
print(path)
```

Import datasets

```python
data = pd.read_csv(f'{path}/data/btw_treatment.csv', encoding = 'UTF-8-SIG', sep=',')
#data = data.set_index(['AGS', 'year'])
data.head(10)
```

Split data in Erst- and Zweitstimme

```python
erst = data[data['Erststimme'] == 1]
zweit = data[data['Zweitstimme'] == 1]
erst.head(10)
```

Calculate average vote shares in all years by treated 50

```python
erst = erst[erst['year'] < 2017]
erst = erst[erst['year'] > 2005]
erst.head()
```

```python
# DiD: y = b0 + b1*D + d0*T + d1*(D*T)
y = erst[['Union']].to_numpy()
x = erst[['treated_50', 'post', 'treated_50*post']].to_numpy()
model = LinearRegression().fit(x, y)
print(f"coefficients: {model.coef_}")
```

```python
y = erst[['SPD']].to_numpy()
x = erst[['treated_50', 'post', 'treated_50*post']].to_numpy()
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
print(results.summary(xname=["const", "treated_50", "post", "treated_50*post"]))
```
