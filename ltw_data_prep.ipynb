Add year column and state number to dataframes

```python
dfs = [by08, by13, by18, he09, he13, he18, nrw10, nrw12, nrw17, ns08, ns13, ns17, th09, th14, th19]
years = [2008, 2013, 2018, 2009, 2013, 2018, 2010, 2012, 2017, 2008, 2013, 2017, 2009, 2014, 2019]
states = ['Bavaria', 'Bavaria', 'Bavaria', 'Hesse', 'Hesse', 'Hesse', 'NRW', 'NRW', 'NRW', 'Lower-Saxony', 'Lower-Saxony', 'Lower-Saxony',  'Thuringia', 'Thuringia', 'Thuringia']
for df, year, state in zip(dfs, years, states):
    df.insert(loc=0, column='year', value=year)
    df.insert(loc=0, column='Land', value=state)
```

Concatenate dataframes

```python
dfs = [by08, by13, by18, he09, he13, he18, nrw10, nrw12, nrw17, ns08, ns13, ns17, th09, th14, th19]
ltw = pd.concat(dfs, sort=False)
```

Fix the AGS

```python
ltw['AGS'] = np.where(ltw['AGS'].str.len() < 6, ltw['AGS'] + '000', ltw['AGS']) # add trailing 000 for kreisfreie Städte (AGS has 4 - 5 digits before, 7 -8 after)
ltw['AGS'] = np.where(ltw['AGS'].str.len() < 8, '0' + ltw['AGS'], ltw['AGS']) # add leading 0 for state ids < 10
```

Drop everything that is not a Gemeinde (e.g. data for States)

```python
#ltw['AGS_len'] = ltw['AGS'].str.len()
ltw = ltw[ltw['AGS'].str.len() == 8]
```

Still there are some duplicates, e.g. Gießen Stadt and Landkreis --> these won't be matched when merging with treatment data

```python
ltw.head()
```

Clean data and sum votes for AfD and Andere

```python
ltw.rename(columns = {'GŸltige Stimmen':'Gültig', 'GrŸne': 'Grüne'}, inplace = True)
andere = ['AfD', 'Sonstige Parteien']
ltw['Andere'] = ltw[andere].sum(axis=1)

ltw.head()
```

Calculate vote shares in percent

```python
parties = ['Union',	'SPD', 'FDP', 'Linke', 'Grüne', 'Andere']
for party in parties:
    ltw.loc[:, party] = (ltw[party] / ltw['Gültig'])*100
ltw.head(10)
```

Add post-treatment indicator

```python
ltw['post_2014'] = np.where(ltw['year'] >= 2014, 1, 0)
ltw['post_2010'] = np.where(ltw['year'] > 2010, 1, 0)
ltw = ltw.sort_values(['AGS', 'year'])
ltw.head(10)
```

Export dataset

```python
ltw = ltw[['AGS', 'Land', 'year', 'post_2010', 'post_2014', 'Union', 'SPD', 'Grüne', 'FDP', 'Linke', 'Andere']]
ltw = ltw.set_index(['AGS', 'year'])
ltw.to_csv(f'{path}/data/ltw.csv', encoding = 'utf-8-sig')
ltw.head()
```

```python
ltw = pd.read_csv(f'{path}/data/ltw.csv', converters={'AGS': str, 'Land': str})
ltw.head()
```

Merge with treatment dataset

```python
treatment = pd.read_csv(f'{path}/data/treatment.csv', converters = {'AGS': str})
#data = btw.join(treatment.set_index(['AGS']))
data = pd.merge(ltw, treatment, on='AGS', how='outer', indicator=True)
data.head(10)
```

Inspect merge results

```python
print(data['_merge'].describe())
```

Clean data

```python
data = data[data['_merge'] == 'both']
data = data.set_index(['AGS', 'GEN', 'year'])
data = data.drop(columns=['_merge'])
data.head(10)
```

Drop missing values

```python
data = data.dropna(axis='rows', how='any', subset=['Union',	'SPD', 'FDP', 'Linke', 'Grüne', 'Andere'])
```

Export dataset

```python
data.to_csv(f'{path}/data/ltw_treatment.csv', encoding = 'utf-8-sig')
```
