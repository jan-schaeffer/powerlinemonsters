First, import needed modules

```python
import numpy as np
import pandas as pd
import os
```

Initalize variables

```python
path = os.getcwd()
print(path)
```

Import dataset

```python
ltw = pd.read_csv(f'{path}/data/ltw.csv', converters={'AGS': str, 'Land': str})
ltw.head()
```

Merge with treatment dataset

```python
treatment = pd.read_csv(f'{path}/data/treatment.csv', converters = {'AGS': str})
#data = btw.join(treatment.set_index(['AGS']))
data = pd.merge(ltw, treatment, on='AGS', how='outer', indicator=True)
data.head(10)
```

Inspect merge results

```python
print(data['_merge'].describe())
```

Clean data

```python
data = data[data['_merge'] == 'both']
data = data.set_index(['AGS', 'GEN', 'year'])
data = data.drop(columns=['_merge'])
data.head(10)
```

Interact treatment dummies with post-treatment indicator

```python
dummies = ['treated_0', 'treated_10', 'treated_20', 'treated_30', 'treated_50', 'treated_100']
for dummy in dummies:
    data.insert(loc=len(data.columns), column=f'{dummy}*post_2014', value=(data[dummy]*data['post_2014']))
data.head(20)
```

Drop missing values

```python
data = data.dropna(axis='rows', how='any', subset=['Union',	'SPD', 'FDP', 'Linke', 'Grüne', 'Andere'])
```

Export dataset

```python
data.to_csv(f'{path}/data/ltw_treatment.csv', encoding = 'utf-8-sig')
```
